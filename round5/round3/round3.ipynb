```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
round1_dayn2 = pd.read_csv("trades_round_3_day_0_wn.csv", sep=';')
round1_dayn1 = pd.read_csv("trades_round_3_day_1_wn.csv", sep=';')
round1_day0 = pd.read_csv("trades_round_3_day_2_wn.csv", sep=';')

round1_dayn1["timestamp"] += 1000000
round1_day0["timestamp"] += 2000000

all_trades_data = pd.concat([round1_dayn2, round1_dayn1, round1_day0])

round1_dayn2 = pd.read_csv("prices_round_3_day_0.csv", sep=';')
round1_dayn1 = pd.read_csv("prices_round_3_day_1.csv", sep=';')
round1_day0 = pd.read_csv("prices_round_3_day_2.csv", sep=';')

round1_dayn1["timestamp"] += 1000000
round1_day0["timestamp"] += 2000000

all_prices_data = pd.concat([round1_dayn2, round1_dayn1, round1_day0])
all_trades_data
```

```python
baskets_trades = all_trades_data[all_trades_data["symbol"] == "GIFT_BASKET"]
strawbs_trades = all_trades_data[all_trades_data["symbol"] == "STRAWBERRIES"]
choc_trades = all_trades_data[all_trades_data["symbol"] == "CHOCOLATE"]
rose_trades = all_trades_data[all_trades_data["symbol"] == "ROSES"]

baskets_prices = all_prices_data[all_prices_data["product"] == "GIFT_BASKET"]
strawbs_prices = all_prices_data[all_prices_data["product"] == "STRAWBERRIES"]
choc_prices = all_prices_data[all_prices_data["product"] == "CHOCOLATE"]
rose_prices = all_prices_data[all_prices_data["product"] == "ROSES"]
```

```python
traders = ["Valentina", "Vinnie", "Vladimir", "Vivian", "Celeste", "Colin", "Carlos", "Camilla", "Pablo", "Penelope", "Percy", "Petunia", "Ruby", "Remy", "Rhianna", "Raj", "Amelia", "Adam", "Alina", "Amir"]
for trader in traders:
    timestamps = baskets_trades[baskets_trades["seller"] == trader]["timestamp"]
    starfruit_trader_sells = []
    for i, row in baskets_prices.iterrows():
        if row["timestamp"] in timestamps.values:
            starfruit_trader_sells.append(row)
    
    sells_timestamps = [x["timestamp"] for x in starfruit_trader_sells]
    sells_prices = [x["mid_price"] for x in starfruit_trader_sells]

    timestamps = baskets_trades[baskets_trades["buyer"] == trader]["timestamp"]
    starfruit_trader_buys = []
    for i, row in baskets_prices.iterrows():
        if row["timestamp"] in timestamps.values:
            starfruit_trader_buys.append(row)
    
    buys_timestamps = [x["timestamp"] for x in starfruit_trader_buys]
    buys_prices = [x["mid_price"] for x in starfruit_trader_buys]

    if len(buys_timestamps) == 0 and len(sells_timestamps) == 0:
        continue
    
    fig, ax = plt.subplots(figsize=(20,10))
    ax.set_title(f"{trader}")
    ax.plot(baskets_prices["timestamp"], baskets_prices["mid_price"])
    ax.scatter(sells_timestamps, sells_prices, c = "r", s=10, marker = "v", zorder=10)
    ax.scatter(buys_timestamps, buys_prices, c = "g", s = 10, marker="^", zorder=9)
```
